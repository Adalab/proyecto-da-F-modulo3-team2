```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Imputación de nulos usando métodos avanzados estadísticos
# -----------------------------------------------------------------------
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

# Librerías de visualización
# -----------------------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_copy = pd.read_csv("df_copy.csv", index_col= 0)
```

```python
##DEMOGRÁFICOS
fig, axes = plt.subplots(4, 2, figsize=(16, 12))
# Gráfico 1
sns.countplot(x='Date Birth', hue='Attrition', data=df_copy, color="violet", ax=axes[0, 0])
axes[0, 0].set_title('Relación entre A y DB')
axes[0, 0].tick_params(axis='x', rotation=90)
# Gráfico 2
sns.barplot(x='Attrition', y='Total Working Years', data=df_copy, color="violet", ax=axes[0, 1])
axes[0, 1].set_title('Relación entre A y TWY')
# Gráfico 3
sns.countplot(x='Gender', hue='Attrition', data=df_copy, color="violet", ax=axes[1, 0])
axes[1, 0].set_title('Relación entre A y G')
# Gráfico 4
sns.countplot(x='Marital Status', hue='Attrition', data=df_copy, color="violet", ax=axes[1, 1])
axes[1, 1].set_title('Relación entre A y MS')
# Gráfico 5
sns.countplot(x='Education', hue='Attrition', data=df_copy, color="violet", ax=axes[2, 0])
axes[2, 0].set_title('Relación entre A y E')
# Gráfico 6
sns.countplot(x='Education Field', hue='Attrition', data=df_copy, color="violet", ax=axes[2, 1])
axes[2, 1].set_title('Relación entre A y EF')
axes[2, 1].tick_params(axis='x', rotation=90)
# Gráfico 7
sns.barplot(x='Attrition', y='Num Companies Worked', data=df_copy, color="violet", ax=axes[3, 0])
axes[3, 0].set_title('Relación entre A y NCW')
# Gráfico 8
sns.barplot(x='Attrition', y='Distance From Home', data=df_copy, color="violet", ax=axes[3, 1])
axes[3, 1].set_title('Relación entre A y DFM')
# Ajustar el diseño
plt.tight_layout()
# Mostrar gráficos
plt.show()
```
